```python
import pandas as pd

# Cargar el archivo de datos desde la carpeta
file_path = './ingresos-1.txt'  # Asegúrate de que el archivo esté en esta carpeta
data = pd.read_csv(file_path, sep="\t")

# Vista rápida de los datos
print(data.head())
print(data.info())
```

```python
# Descripción estadística
print(data.describe())

# Verificar valores nulos
print(data.isnull().sum())
```

```python
# Imputar valores faltantes con la mediana
data['educacion_madre'].fillna(data['educacion_madre'].median(), inplace=True)
data['educacion_padre'].fillna(data['educacion_padre'].median(), inplace=True)
data['peso'].fillna(data['peso'].median(), inplace=True)
# Imputar valores categóricos con el más frecuente (moda)
data['fumador'].fillna(data['fumador'].mode()[0], inplace=True)
data['tenso'].fillna(data['tenso'].mode()[0], inplace=True)
data['malhumorado'].fillna(data['malhumorado'].mode()[0], inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribución de ingresos
plt.figure(figsize=(8, 6))
sns.histplot(data['ingresos'], kde=True, bins=30)
plt.title('Distribución de los Ingresos')
plt.xlabel('Ingresos (USD)')
plt.ylabel('Frecuencia')
plt.show()

# Boxplot para explorar la relación entre género y ingresos
plt.figure(figsize=(8, 6))
sns.boxplot(x='genero', y='ingresos', data=data)
plt.title('Distribución de Ingresos por Género')
plt.xlabel('Género')
plt.ylabel('Ingresos (USD)')
plt.show()
```

### Análisis de la Distribución de los Ingresos

#### Distribución General de los Ingresos
El primer gráfico muestra la distribución de los ingresos anuales en el dataset. Podemos observar lo siguiente:
- La distribución es altamente **sesgada a la derecha**, lo que significa que la mayoría de los encuestados tienen ingresos bajos (entre $0 y $50,000).
- Hay **valores atípicos (outliers)** que representan ingresos muy altos, alcanzando más de $400,000, aunque estos casos son raros.

Esta asimetría sugiere la necesidad de transformar los datos, como aplicar una transformación logarítmica, para normalizar la distribución y facilitar el análisis.

---

#### Distribución de Ingresos por Género
El segundo gráfico es un boxplot que compara los ingresos entre los géneros. Algunos puntos clave:
- La mayoría de los ingresos están concentrados en valores bajos para ambos géneros.
- Los valores atípicos (puntos fuera de los bigotes del boxplot) son evidentes en ambos géneros y deben ser tratados con cuidado.
- Aunque el rango intercuartil (IQR) es similar para ambos géneros, sería interesante analizar si hay diferencias significativas en los ingresos promedio utilizando pruebas estadísticas.

---

#### Siguientes Pasos
1. **Manejo de valores atípicos**: Se podrían eliminar o transformar para reducir su impacto en el análisis.
2. **Transformación logarítmica**: Aplicar una transformación logarítmica podría ayudar a normalizar la distribución de los ingresos.
3. **Relación con otras variables**: Es importante analizar cómo los ingresos varían según otras variables, como años de educación, edad, y etnia.




```python
# Eliminar outliers
data = data[data['ingresos'] < 100000]
```

```python
import numpy as np
data['ingresos_log'] = np.log1p(data['ingresos'])  # log(ingresos + 1)
```

```python
sns.histplot(data['ingresos_log'], kde=True)
plt.title("Distribución de los Ingresos (Transformación Logarítmica)")
plt.show()
```

#### Observaciones
1. La transformación logarítmica ha reducido significativamente el sesgo hacia la derecha presente en la distribución original de los ingresos.
2. Ahora los ingresos están más concentrados alrededor de un rango intermedio (valores logarítmicos entre 8 y 10), lo que facilita el análisis.
3. Todavía se observa un pico en valores bajos (cercanos a 0 en la escala logarítmica), lo cual corresponde a ingresos muy bajos o iguales a cero en los datos originales.


```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Imputar valores faltantes en las columnas numéricas
data_numeric = data.select_dtypes(include=['float64', 'int64']).drop(columns=['ingresos'])  # Excluir "ingresos" original
data_numeric.fillna(data_numeric.median(), inplace=True)  # Imputar con la mediana

# Escalar los datos
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_numeric)

# Aplicar PCA para reducir las dimensiones
pca = PCA(n_components=2)  # Reducir a 2 componentes principales
principal_components = pca.fit_transform(data_scaled)

# Crear un DataFrame con los componentes principales
pca_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
print(pca_df.head())

# Visualizar los resultados del PCA
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.scatterplot(x='PC1', y='PC2', data=pca_df, hue=data['genero'], palette='viridis')  # Usar la columna 'genero' como color
plt.title('PCA - Componentes Principales')
plt.xlabel('Componente Principal 1 (PC1)')
plt.ylabel('Componente Principal 2 (PC2)')
plt.show()
```

### Reducción de Dimensionalidad con PCA

#### Descripción
El análisis de componentes principales (PCA) se utilizó para reducir la dimensionalidad del conjunto de datos a 2 componentes principales. Esto permite visualizar los datos en un espacio de menor dimensión mientras se conserva la mayor cantidad posible de variabilidad original.

---

#### Observaciones del Gráfico
1. El gráfico de dispersión muestra las observaciones proyectadas en el espacio definido por los dos componentes principales (`PC1` y `PC2`).
2. Cada punto representa un individuo, y los colores corresponden a la variable categórica `genero`:
   - Azul: Masculino.
   - Verde: Femenino.
3. Existe una superposición significativa entre ambos géneros, lo que sugiere que las diferencias entre ellos, en términos de las variables incluidas en el análisis, no son claras en este espacio reducido.

---

#### Explicación de los Componentes Principales
1. **Componente Principal 1 (PC1):**
   - Este eje captura la mayor cantidad de variabilidad en los datos.
   - Tiene la mayor contribución de las variables originales escaladas.
2. **Componente Principal 2 (PC2):**
   - Este eje captura la segunda mayor cantidad de variabilidad, ortogonal al primero.
   - Representa una perspectiva complementaria de las relaciones entre las variables originales.

---

#### Utilidad del PCA
- La reducción de dimensionalidad facilita la visualización y el análisis exploratorio de los datos.
- Permite identificar patrones, agrupaciones o tendencias que no eran evidentes en el espacio original de alta dimensionalidad.
- Estos componentes principales pueden ser utilizados en análisis posteriores o como insumos para modelos predictivos.
